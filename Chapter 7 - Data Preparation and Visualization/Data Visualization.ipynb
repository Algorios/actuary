```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns; sns.set()
```

```python
%matplotlib inline
```

# Bar Chart

```python
plt.bar([1, 2, 3], [4, 5, 6])

plt.xlabel('x vals')

plt.ylabel('y vals')

plt.title('My Bar Chart');
```

# Boxplot

```python
plt.boxplot([1, 2, 3, 4])

plt.xlabel('My Variable')

plt.ylabel('Values')

plt.title('Boxplot of Variable');
```

# Histogram

```python
plt.hist([1, 2, 3, 4])

plt.xlabel('x values')

plt.ylabel('Frequency')

plt.title('Histogram Example');
```

# Scatter Plot

```python
plt.scatter([1, 2, 3, 4], [5, 6, 7, 8])

plt.xlabel('x vals')

plt.ylabel('y vals')

plt.title('My Scatter');
```
